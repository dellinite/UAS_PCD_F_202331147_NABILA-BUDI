```python
#202331147_Nabila
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#202331147_Nabila
img = cv2.imread("gbk.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#202331147_Nabila
# Median Filtering
median_filtered = cv2.medianBlur(img_rgb, 3)  # Kernel size 3x3

plt.figure(figsize=(10, 6))  

# Tampilkan
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img_rgb)
plt.axis("on")

plt.subplot(1, 2, 2)
plt.title("After filter median")
plt.imshow(median_filtered)
plt.axis("on")

plt.show()
```

```python
#202331147_Nabila
# Ubah ke grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Padding image
padded = np.pad(gray, ((1, 1), (1, 1)), mode='constant', constant_values=0)

# Buat array kosong untuk hasil
mean_filtered = np.zeros_like(gray)

# Loop untuk menghitung mean 3x3
for i in range(mean_filtered.shape[0]):
    for j in range(mean_filtered.shape[1]):
        region = padded[i:i+3, j:j+3]
        mean_val = np.mean(region)
        mean_filtered[i, j] = int(mean_val)

plt.figure(figsize=(10, 6))  # Lebar x Tinggi dalam inci

# Tampilkan
plt.subplot(1, 2, 1)
plt.title("Original Grayscale")
plt.imshow(gray, cmap='gray')
plt.axis("on")

plt.subplot(1, 2, 2)
plt.title("Mean Filtered Image")
plt.imshow(mean_filtered, cmap='gray')
plt.axis("on")

plt.show()
```
