```python
#202331147_Nabila
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#202331147_Nabila
img = cv2.imread('1.jpg') 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
original = img_rgb.copy()
```

```python
#202331147_Nabila
(h, w) = img.shape[:2]
center = (w // 2, h // 2)
M_rotate = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img_rgb, M_rotate, (w, h))

# Resized (ukuran lebih kecil)
resized = cv2.resize(img_rgb, (w // 6, h))

# Cropped (potong tengah)
crop_w = w // 4
crop_h = h // 4
start_x = w // 2 - crop_w // 2
start_y = h // 2 - crop_h // 2
end_x = start_x + crop_w
end_y = start_y + crop_h
cropped = img_rgb[start_y:end_y, start_x:end_x]

# Flipped (horizontal)
flipped = cv2.flip(img_rgb, 1)

#  Translated (geser ke kanan & bawah)
M_trans = np.float32([[1, 0, 100], [0, 1, 50]])
translated = cv2.warpAffine(img_rgb, M_trans, (w, h))

# Menampilkan semua gambar dalam satu figure
titles = ['Original', 'Rotated 45', 'Resized', 'Cropped', 'Flipped', 'Translated']
images = [original, rotated, resized, cropped, flipped, translated]

plt.figure(figsize=(15, 8))
for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
